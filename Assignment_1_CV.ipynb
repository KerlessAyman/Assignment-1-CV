```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python

url = "/fruits.csv"

df = pd.read_csv(url)

df.head()
```

#Bar Chart

```python
plt.figure(figsize=(8,5))
plt.bar(df['Fruit'], df['Quantity'])
plt.title('Total Quantities Sold per Product')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

#Line Chart


```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Price'], marker='o')
plt.title('Prices Over Months')
plt.xlabel('Month')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

#Pie Chart


```python
df['Sales'] = df['Quantity'] * df['Price']

plt.figure(figsize=(7,7))
plt.pie(df['Sales'], labels=df['Fruit'], autopct='%1.1f%%', startangle=90)
plt.title('Sales Share by Product')
plt.show()
```

##Plotly Express

#Bar Chart

```python
fig = px.bar(df, x='Fruit', y='Quantity', title='Total Quantities Sold per Product')
fig.show()
```

#Line Chart

```python
fig = px.line(df, x='Month', y='Price', title='Prices Over Months', markers=True)
fig.show()
```

#Pie Chart

```python
fig = px.pie(df, names='Fruit', values='Sales', title='Sales Share by Product')
fig.show()
```

# Step 1. Import necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# Step 2. Load the image

```python
img = cv2.imread('/apple.jpg')
```

# Step 3. Print its matrix representation

```python

print("Matrix representation of the colored image:\n")
print(img)
```

## Step 4. Convert it to grayscale

```python

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

# Step 5. Print its grayscale matrix

```python
print("\nMatrix representation of the grayscale image:\n")
print(gray)
```

# Step 6. Display both colored and grayscale images

```python

plt.figure(figsize=(10,5))


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Original colored image
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Colored Image")
plt.axis("off")

# Grayscale image
plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
